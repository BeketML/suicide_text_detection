```python
import re
import html
import unicodedata
import string
from bs4 import BeautifulSoup
import emoji

URL_PATTERN = re.compile(
    r"(https?://\S+|www\.\S+)",
    flags=re.IGNORECASE
)
EMAIL_PATTERN = re.compile(
    r"\b[\w\.-]+@[\w\.-]+\.\w+\b",
    flags=re.IGNORECASE
)
MENTION_PATTERN = re.compile(
    r"@\w+",
    flags=re.UNICODE
)
MULTI_SPACE_PATTERN = re.compile(r"\s+")

def clean_text_combined(text):
    """
    All-in-one text cleaning: HTML/Unicode, demojize, lowercase, 
    replace URLs/emails/mentions, remove digits, remove special chars, 
    normalize spaces, remove HTML tags.
    """
    import pandas as pd
    # If text is null or not string/pd.NA, return empty string
    if text is None or (hasattr(pd, 'isnull') and pd.isnull(text)):
        return ""
    text = str(text)
    # Unescape HTML entities
    text = html.unescape(text)
    # Normalize unicode
    text = unicodedata.normalize("NFKC", text)
    # Convert emojis to text (e.g. ":smiling_face:")
    text = emoji.demojize(text)
    # Lowercase all
    text = text.lower()
    # Substitute URLs and emails
    text = URL_PATTERN.sub(" <url> ", text)
    text = EMAIL_PATTERN.sub(" <email> ", text)
    # Substitute @mentions
    text = MENTION_PATTERN.sub(" <user> ", text)
    # Remove digits
    text = re.sub(r'\d+', '', text)
    # Remove HTML tags (safety, after above)
    text = BeautifulSoup(text, "html.parser").get_text()
    # Remove non-word/space/[.,!?…\-–—] punctuation but preserve important punct
    text = re.sub(r"[^\w\s.,!?…\-–—]", " ", text, flags=re.UNICODE)
    # Remove all (other) punctuation except allowed
    allowed_punct = ".,!?…\-–—"
    clean_chars = []
    for c in text:
        if c.isalnum() or c.isspace() or c in allowed_punct:
            clean_chars.append(c)
        # else drop it
    text = ''.join(clean_chars)
    # Normalize whitespace
    text = MULTI_SPACE_PATTERN.sub(" ", text).strip()
    return text
```

### Language Detection

This section adds a `language` column identifying the language of each text entry using kaznlp library. The library can detect Russian, Kazakh, and other languages.


Detect language using kaznlp: https://github.com/makazhan/kaznlp.git

We add the `language` column only for EDA and data analysis purposes.
We use this specific kaznlp library because it provides the best performance for detecting Kazakh and Russian languages.


```python
import os
import sys

# Добавляем текущую директорию (где находится ноутбук) в sys.path
# Это позволяет импортировать helpers.kaznlp.kaznlp...
current_dir = os.getcwd()
if current_dir not in sys.path:
    sys.path.insert(0, current_dir)

# Импортируем модули kaznlp используя полный путь как в других ноутбуках
from helpers.kaznlp.kaznlp.lid.lidnb import LidNB
from helpers.kaznlp.kaznlp.tokenization.tokrex import TokenizeRex

# Инициализируем токенизатор и определитель языка один раз для оптимизации
tokrex = TokenizeRex()
# Путь к модели символов
kaznlp_base_path = os.path.join(current_dir, 'helpers', 'kaznlp')
cmdl = os.path.join(kaznlp_base_path, 'kaznlp', 'lid', 'char.mdl')
landetector = LidNB(word_mdl=None, char_mdl=cmdl)

def detect_language(text):
    """
    Определяет язык текста используя kaznlp библиотеку.
    Возвращает: 'kazakh', 'russian', 'english', или 'other'
    """
    if pd.isna(text) or text is None or str(text).strip() == '':
        return 'other'
    
    try:
        # Токенизация и предсказание языка
        tokens = tokrex.tokenize(str(text), lower=True)[0]
        
        if not tokens:
            return 'other'
        
        lang = landetector.predict(tokens)
        
        # Нормализация названий языков
        lang_lower = lang.lower() if lang else 'other'
        if 'kazakh' in lang_lower or 'kk' in lang_lower:
            return 'kazakh'
        elif 'russian' in lang_lower or 'ru' in lang_lower or 'rus' in lang_lower:
            return 'russian'
        elif 'english' in lang_lower or 'en' in lang_lower or 'eng' in lang_lower:
            return 'english'
        else:
            return 'other'
    except Exception as e:
        # Не выводим ошибку для каждого текста, чтобы не засорять вывод
        return 'other'
```

```python
# Применяем определение языка к колонке text
# Это может занять некоторое время для большого датасета
from tqdm import tqdm
tqdm.pandas(desc="Определение языка")

print("Начало определения языка...")
print(f"Всего строк для обработки: {len(all_data_cleaned)}")

# Применяем функцию detect_language к колонке text с прогресс-баром
all_data_cleaned["language"] = all_data_cleaned["clean_text"].progress_apply(detect_language)

print("\nОпределение языка завершено!")
print("\nРаспределение языков:")
print(all_data_cleaned["language"].value_counts())
```
